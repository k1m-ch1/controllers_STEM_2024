```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
first_experiment_df = pd.read_csv("../sources/pwm_voltage_rpm_data.csv", sep='\t')
```

```python
# Create figure and axes objects
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 8))

fig.suptitle('Experiment initial voltage 12.86V, final voltage 12.55V')

# Plot on the first subplot
axes[0].plot(first_experiment_df['PWM '], first_experiment_df['voltage '])


axes[0].set_xlim(left=0)
axes[0].set_ylim(bottom=0)

axes[0].set_ylabel('Voltage accross the motor')
axes[0].set_xlabel('PWM(8-bit)')

axes[1].plot(first_experiment_df['PWM '], first_experiment_df['count per second'])
axes[1].set_xlim(left=0)
axes[1].set_ylim(bottom=0)

axes[1].set_ylabel('count per second')
axes[1].set_xlabel('PWM(8-bit)')


# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
